```python
import numpy as np
import matplotlib.pyplot as plt

graph = np.random.binomial(100, .5, 100)
plt.hist(graph, bins=20, color='Darkred')
plt.axvline(graph.mean(), color='pink', linestyle='solid', linewidth=2)
plt.axvline(graph.mean() + graph.std(), color='pink', linestyle='dashed', linewidth=2)
plt.axvline(graph.mean() - graph.std(), color='pink', linestyle='dashed', linewidth=2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
distributed1 = np.random.normal(5, .5, 100)
distributed2 = np.random.normal(10, 1, 100)
distributed3 = distributed1 + distributed2

plt.hist(distributed3, bins=20, color='c')
plt.axvline(distributed3.mean(), color='pink', linestyle='solid', linewidth=2)
plt.axvline(distributed3.mean() + distributed3.std(), color='pink', linestyle='dashed', linewidth=2)
plt.axvline(distributed3.mean() - distributed3.std(), color='pink', linestyle='dashed', linewidth=2)
plt.show()
```
