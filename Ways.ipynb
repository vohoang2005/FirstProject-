```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

df = pd.read_csv("purchases.csv")
bf = pd.read_csv("titanic.csv")
df.head()
```

## Answer to number 1

## Since this plot only runs off of 1 variable, you can't really gather any information on it but where it starts, medium and ends. Bad plot.

```python
sns.boxplot(x=df["ad_views"])
```

## A bar graph, which only shows it start at zero and stopped around 37. Plot does not give much detail. Bad graph

```python
sns.barplot(x=df["ad_views"])
```

## Answer to number 2

## This graph give better details than Question 1 does. But the accuracy of the graph is very bad. We can't gather alot of details on this graph.

```python
sns.jointplot(x="ad_views", y="items_purchased", data=df, size=4)
```

## This graph shows the views count and number of purchase. It a decent graph that help us realize that views does not mean items purchased.

```python
sns.factorplot(x="ad_views", y="items_purchased", data=df, size=4)
```

## This violin graph does not work well with 2 continous variable. Should not be used in this case.

```python
sns.violinplot(x="ad_views", y="items_purchased", data=df, size=4)
```

## Answer to number 3

## This box graph is decent, it shows which Country have the least, medium and most view counts.

```python
sns.boxplot(x="country", y ="ad_views", data=df)
```

## A more decent violin plot from above. It show how much people watch the ads and which area they view the most, but in a very different way.

```python
sns.violinplot(x="country", y="ad_views", data=df, size=6)
```

## Very basic plot, show the medium of the 2 ad views by country. A good graph for simplicity people to look at.

```python
sns.set(style="darkgrid")
sns.factorplot(x="country", y="ad_views", data=df, size=6, palette="pastel")
```

```python
sns.barplot(x="country", y="ad_views", data=df)
```

```python
sns.stripplot(x="country", y="ad_views", data=df)
```
