```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df = pd.read_csv("titanic.csv")


df.Survived.value_counts().plot(kind="pie", autopct='%1.1f%%')
plt.legend(labels=['dead', 'survived'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("titanic.csv")
df = pd.get_dummies(df, columns=['Sex']) # look up what get_dummies() does to the column passed in

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(df[df['Sex_male'] == 1.0]['Age'].dropna(), bins=10, color='purple') # look up what dropna() does
plt.ylabel('Male')
plt.xlabel('Age Group')
plt.title('Male that went Aboard')

plt.subplot(1, 2, 2)
plt.hist(df[df['Sex_female'] == 1]['Age'].dropna(), bins=10, color='green')
plt.ylabel('Female')
plt.xlabel('Age Group')
plt.title('Female that went Aboard')

plt.show()
```
